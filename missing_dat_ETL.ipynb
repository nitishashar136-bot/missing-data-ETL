**Q1. What are the most common reasons for missing data in ETL pipelines?**

- 1. Data entry mistakes

2. System errors during data collection

3. Data not available at source

4. Integration issues from multiple systems

5. Network/server failure during extraction

**Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?**

- It reduces dataset size

It may remove important information

It can create bias in analysis

It affects business decisions

**Q3. Explain the difference between:
Listwise deletion
Column deletion
Also mention one scenario where each is appropriate.**

- # Listwise Deletion

Remove entire row if any value is missing.

Appropriate when:
Very few rows have missing data.

# Column Deletion

Remove entire column if most values are missing.

Appropriate when:
Column has too many missing values and is not important.

**Q4. Why is median imputation preferred over mean imputation for skewed data such as income?**
- Income data usually has extreme values (very high earners).
Mean gets affected by extreme values.
Median gives more accurate middle value.
So median is better for skewed data.

**Q5. What is forward fill and in what type of dataset is it most useful?**
- Forward fill replaces missing value with previous value.

Useful in:
Time-series data (sales, stock prices, daily records).

**Q6. Why should flagging missing values be done before imputation in an ETL workflow?**

- Because after filling values, we lose information about what was originally missing.

Flagging helps:
Track missing data
Analyze patterns
Improve business insights
